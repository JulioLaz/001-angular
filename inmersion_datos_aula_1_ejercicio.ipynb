#Calcular el promedio de area para cada uno de los barrios y hacer grafico con top 10

```python
import pandas as pd
```

##Abrir base de datos

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/inmuebles_bogota.csv')
data.head(10)
```

```python
data.shape
```

#RENOMBRAR COLUMNAS

```python
columnas_dict={'Baños':'Banos', 'Área':'Area'}
data=data.rename(columns=columnas_dict)
data.head()
```

##Crear lista con nombres de los barrios:

```python
data.Barrio.value_counts()
serie_datos=data.Barrio.value_counts()
list_barrios = serie_datos.index.values
list_barrios
```

```python
# Cantidad de Barrios
len(list_barrios)
```

##Calculo el promedio de area por barrio y crear lista de pormedios de areas

```python
#Calculo n: cantidad de barrios 
n = len(list_barrios)-1
lista_prom = [] # Crear una lista vacía
#ciclo for de 0 a n:
for i in range(n+1):
  area_promedio= data[(data.Barrio==serie_datos.index[i])].Area.mean().round(1)
  #Lista con los promedios por barrio:
  lista_prom.append(area_promedio)
```

```python
lista_prom
```

##Crear DataFrame con barrios y promedios de areas:

```python
# Crea un diccionario con los datos

#Area_Promedio=lista_prom
datos = {'Area Promedio':lista_prom}

# Crea el dataframe a partir del diccionario y coloco por index la lista de barrios:list_barrios
df_barrios_area = pd.DataFrame(datos,index=list_barrios)

df_barrios_area.index.name='Barrio'
df_barrios_area
```

##Ordenar en orden descendiente

```python
df_area_prom_x_barrio = df_barrios_area.sort_values(by='Area Promedio', ascending=False)
df_area_prom_x_barrio
```
